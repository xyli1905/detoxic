```python
def pre_AUC(fname):
    fdir = "/Users/xyli1905/Projects/NLP/detoxic/results/threshold"
    fpath = os.path.join(fdir, fname)
    data = pd.read_csv(fpath, index_col=0)
    data_val = data.values
    # print(R1D1d0_val.shape)
    data_recall = np.hstack((1.,data_val[6]))
    data_fallout = np.hstack((1.,data_val[8]))
    return data_fallout, data_recall
```

```python
def plot_2_AUC(fallout1, recall1, fallout2, recall2):
    plt.plot(fallout1, recall1, 'r')
    plt.fill_between(fallout1, recall1, facecolor='red', alpha=0.5)
    plt.plot(fallout2, recall2, 'b')
    plt.fill_between(fallout2, recall2, facecolor='blue', alpha=0.5)
    aax = np.array([0,1])
    aay = np.array([0,1])
    plt.plot(aax,aay,'--', c='black')
    plt.fill(aax,aay, facecolor='blue', alpha=0.5)

    ax = plt.gca()
    fig = plt.gcf()
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_aspect('equal', adjustable='box')
    fig.set_size_inches(7,7)

    plt.show()
```

```python
def plot_3_AUC(fallout1, recall1, fallout2, recall2, fallout3, recall3):
    plt.plot(fallout1, recall1, 'r', label = 'LSTM')
#     plt.fill_between(fallout1, recall1, facecolor='red', alpha=0.7)
    plt.plot(fallout2, recall2, 'b', label = 'EmbLR')
#     plt.fill_between(fallout2, recall2, facecolor='#5fff0c', alpha=0.7)
    plt.plot(fallout3, recall3, 'g', label = 'BoW')
#     plt.fill_between(fallout3, recall3, facecolor='#1956ff', alpha=0.7)

    aax = np.array([0,1])
    aay = np.array([0,1])
    plt.plot(aax,aay,'--', c='black')
    plt.fill(aax,aay, facecolor='blue', alpha=0.5)

    plt.grid(True)
    plt.title('AUC-ROC curves for different models', fontsize = 20, y=1.04)
    plt.xlabel('False Positive Rate (Fallout)', fontsize=18)
    plt.ylabel('Ture Positive Rate (Recall)', fontsize=18)

    fig = plt.gcf()
    fig.set_size_inches(7,7)

    ax = plt.gca()
    ax.set_xlim(0,1)
    ax.set_ylim(0,1)
    ax.set_aspect('equal', adjustable='box')
    ax.set_facecolor('#edf0f8')
    ax.grid(color='w', linestyle='-.', linewidth=1)
    ax.legend(shadow=True, fontsize='large')
    
    ax.tick_params(labelsize=13)
    
    plt.savefig('/Users/xyli1905/Desktop/AUC.eps', format='eps', dpi=1000)

    plt.show()
```

```python
def pre_table(fname, col):
    fdir = "/Users/xyli1905/Projects/NLP/detoxic/results/baseline"
    fpath = os.path.join(fdir, fname)
    data = pd.read_csv(fpath, index_col=0)
    data_val = data.values
    # print(R1D1d0_val.shape)
    data_opt = data_val[5:, col-1]
    return data_opt.reshape(1,-1)
```
